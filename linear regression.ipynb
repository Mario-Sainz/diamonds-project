```python
import pandas as pd
```

```python
diamonds =  pd.read_csv('train.csv')
```

```python
diamonds.head()
```

```python
diamonds.isna().sum()
```

```python
print(diamonds['cut'].value_counts())
```

```python
print(diamonds['color'].value_counts())
```

```python
print(diamonds['clarity'].value_counts())
```

```python
pd.get_dummies(diamonds, columns=["cut", "color", "clarity"])
```

```python
diamonds = pd.get_dummies(diamonds, columns=["cut", "color", "clarity"], drop_first=True)
```

```python
diamonds.corr()
```

```python
diamonds.info()
```

```python
from sklearn.model_selection import train_test_split
target = "price"
```

```python
X = diamonds.drop(target, axis=1)
y = diamonds[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_test = lr.predict(X_test)
y_pred_test
```

```python
y_pred_train = lr.predict(X_train)
y_pred_train
```

```python
lr.score(X, y)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(
    y_train,
    y_pred_train
)
```

```python
mean_squared_error(
    y_test,
    y_pred_test
)
```
